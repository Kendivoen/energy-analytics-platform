## Preparing Carbon Impact Dashboard Assets

The carbon impact analysis generated several important sustainability metrics and visualizations.

To support dashboard development, the carbon impact dataset is transformed into monthly, annual, and seasonal summary tables that can be loaded directly into the dashboard.

```python
# Ensure datetime format

carbon_df['utc_timestamp'] = pd.to_datetime(
    carbon_df['utc_timestamp']
)

# Create year and month columns

carbon_df['year'] = carbon_df['utc_timestamp'].dt.year

carbon_df['year_month'] = (
    carbon_df['utc_timestamp']
    .dt.to_period('M')
    .astype(str)
)

print("Carbon dataset prepared.")
```

### Monthly Carbon Impact Dataset

Monthly carbon impact data will support trend analysis and time-series visualizations within the dashboard.

```python
monthly_carbon = (
    carbon_df.groupby('year_month')[
        [
            'solar_co2_avoided_tonnes',
            'wind_co2_avoided_tonnes',
            'total_co2_avoided_tonnes'
        ]
    ]
    .sum()
    .reset_index()
)

monthly_carbon.to_csv(
    "../data/dashboard/monthly_carbon_impact.csv",
    index=False
)

monthly_carbon.head()
```

### Annual Carbon Impact Dataset

Annual carbon impact data provides a high-level view of the environmental benefits delivered by renewable electricity generation over time.

This dataset will support annual trend visualizations and sustainability reporting within the dashboard.

```python
# Annual carbon impact dataset

annual_carbon = (
    carbon_df.groupby('year')[
        [
            'solar_co2_avoided_tonnes',
            'wind_co2_avoided_tonnes',
            'total_co2_avoided_tonnes'
        ]
    ]
    .sum()
    .reset_index()
)

# Round values and convert to integers

annual_carbon[
    [
        'solar_co2_avoided_tonnes',
        'wind_co2_avoided_tonnes',
        'total_co2_avoided_tonnes'
    ]
] = (
    annual_carbon[
        [
            'solar_co2_avoided_tonnes',
            'wind_co2_avoided_tonnes',
            'total_co2_avoided_tonnes'
        ]
    ]
    .round(0)
    .astype(int)
)

# Save dashboard asset

annual_carbon.to_csv(
    "../data/dashboard/annual_carbon_impact.csv",
    index=False
)

print("Annual carbon impact dataset saved.")

annual_carbon
```

The annual carbon impact dataset summarizes avoided emissions from solar generation, wind generation, and total renewable generation for each year in the study period.

This dataset will be used to create annual carbon impact charts within the dashboard.

### Seasonal Carbon Impact Dataset

Renewable generation exhibits strong seasonal variability due to changing weather conditions and resource availability.

The seasonal carbon impact dataset summarizes avoided emissions by season and will support comparative visualizations within the dashboard.

```python
# Create season column

carbon_df['month'] = carbon_df['utc_timestamp'].dt.month

season_map = {
    12: 'Winter', 1: 'Winter', 2: 'Winter',
    3: 'Spring', 4: 'Spring', 5: 'Spring',
    6: 'Summer', 7: 'Summer', 8: 'Summer',
    9: 'Autumn', 10: 'Autumn', 11: 'Autumn'
}

carbon_df['season'] = carbon_df['month'].map(season_map)

# Aggregate by season

seasonal_carbon = (
    carbon_df.groupby('season')[
        [
            'solar_co2_avoided_tonnes',
            'wind_co2_avoided_tonnes',
            'total_co2_avoided_tonnes'
        ]
    ]
    .sum()
    .reset_index()
)

# Logical season order

season_order = {
    'Winter': 0,
    'Spring': 1,
    'Summer': 2,
    'Autumn': 3
}

seasonal_carbon['order'] = (
    seasonal_carbon['season']
    .map(season_order)
)

seasonal_carbon = (
    seasonal_carbon
    .sort_values('order')
    .drop(columns='order')
)

# Round values and convert to integers

seasonal_carbon[
    [
        'solar_co2_avoided_tonnes',
        'wind_co2_avoided_tonnes',
        'total_co2_avoided_tonnes'
    ]
] = (
    seasonal_carbon[
        [
            'solar_co2_avoided_tonnes',
            'wind_co2_avoided_tonnes',
            'total_co2_avoided_tonnes'
        ]
    ]
    .round(0)
    .astype(int)
)

# Save dashboard asset

seasonal_carbon.to_csv(
    "../data/dashboard/seasonal_carbon_impact.csv",
    index=False
)

print("Seasonal carbon impact dataset saved.")

seasonal_carbon
```

The seasonal carbon impact dataset captures the complementary behaviour of solar and wind resources throughout the year.

This dataset will support seasonal sustainability visualizations and help communicate the role of renewable resource diversity in reducing carbon emissions.
